## Creating  dataframe with only list

```python
#first we are creating dataframe with the help of list 

lis = [2,4,8,5,2,5,8,5,4,1]

dataframe = pd.DataFrame(lis)
dataframe
```

```python
# we can also creat dataframe with multiple of libaries

lis2 = [[2,4,8,5,2,5,8,5,4,1],[10,40,80,50,20,50,80,50,40,10],[21,41,18,511,211,15,890,534,344,431]]

dataframe2 = pd.DataFrame(lis2 , columns = ["num1","num2","num3","num4","num5","num6","num7","num8","num9","num10"])#columns name

dataframe2
```
